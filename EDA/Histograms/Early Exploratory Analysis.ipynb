# Early Exploratory Data Analysis- Histograms

### Histograms Examples

Histograms are useful in Early Exploratorty Data Anlaysis (EDA), for they allow a quick glance at the data and allow to view its distribution and infer about 

- center (mean, median, mode), 
- shape (symmetric, skewed left, skewed right, uniform), 
- spread (tightly clustered around center, spread out), 
- modality (number of peaks).

#### Center
- Unimodal 
- Bimodal 
- Symmetric 
- Skewed 

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from scipy import stats
from scipy.stats import skewnorm
from scipy.stats import kurtosis
from scipy.stats import skew

#make this example reproducible
np.random.seed(0)

#create normally distributed data
x = np.random.normal(size=1000)

#create histogram of normally distributed
sns.displot(x,color='darkgray')
plt.title("Symmetric, Unimodal, Bell-Shaped", fontsize = 15)
plt.xlabel("Range")
plt.ylabel("Frequency")
plt.savefig("Unimodal.png",bbox_inches='tight')
```

```python
#create trace of normally distributed data
sns.displot(x, kind='kde',color='darkgray')
plt.title("Symmetric, Unimodal, Bell-Shaped", fontsize = 15)
plt.xlabel("Range")
plt.savefig("Unimodal_Trace.png",bbox_inches='tight')
```

```python
#Trace+ Histogram of normally distributed data
sns.displot(x, kde=True,color='darkgray',line_kws={"color":"red"})
plt.title("Symmetric, Unimodal, Bell-Shaped", fontsize = 15)
plt.xlabel("Range")
plt.savefig("Unimodal_plus_Trace.png",bbox_inches='tight')
```

```python
#Trace and CENTER statistics of normally distributed data
mode= stats.mode(x)
sns.displot(x, kind='kde', color='darkgray')
plt.title("Symmetric, Unimodal, Bell-Shaped\n (mean=median)", fontsize = 15)
plt.xlabel("Range")
plt.ylabel("Frequency")
plt.axvline(x.mean(), color='red', label='mean', linewidth=2)
plt.axvline(np.median(x), color='black', label='median', linestyle= 'dashed',linewidth=2)
plt.axvline(mode[0][0], color='blue', label='mode', linestyle= 'dashed',linewidth=2)

plt.text(1.8, 0.32,f'mean:{np.mean(x):.2F}',fontsize="large",fontstyle='normal',fontweight='bold')
plt.text(1.8, 0.30,f'median:{np.median(x):.2F}',fontsize="large",fontstyle='normal',fontweight='bold')
plt.text(1.8,0.28,f'mode:{mode[0][0]:.2F}', fontsize="large", fontstyle='normal', fontweight='bold')
plt.legend()
plt.savefig("Unimodal_Trace_Stats.png",bbox_inches='tight')
```

```python
#create left skewed data distributed data
x= skewnorm.rvs(a= -5, size=1000)  #Skewnorm function (negative number left skew)

#create histogram of left-skewed distributed
sns.displot(x, color='darkgray',line_kws={"color":"red"})
plt.title("Left Skewed", fontsize = 15)
plt.xlabel("Range")
plt.ylabel("Frequency")
plt.savefig("Left_Skewed.png",bbox_inches='tight')
```

```python
#create trace for left-skewed distributed data
sns.displot(x, kind='kde', color='darkgray')
plt.title("Left Skewed", fontsize = 15)
plt.xlabel("Range")
plt.savefig("Left_Skewed_Trace.png",bbox_inches='tight')
```

```python
#Trace+ Histogram of left-skewed distributed data
sns.displot(x, kde=True,color='darkgray',line_kws={"color":"red"})
plt.title("Left Skewed", fontsize = 15)
plt.xlabel("Range")
plt.ylabel("Frequency")
plt.savefig("Left_Skewed_plus_Trace.png",bbox_inches='tight')
```

```python
#Trace and CENTER statistics for left-skewed distributed data
mode= stats.mode(x)
sns.displot(x, kind='kde', color='darkgray')
plt.title("Left Skewed\n (mean < median)", fontsize = 15)
plt.xlabel("Range")
plt.axvline(x.mean(), color='red', label='mean', linewidth=2)
plt.axvline(np.median(x), color='black', label='median', linestyle= 'dashed',linewidth=2)
plt.axvline(mode[0][0], color='blue', label='mode', linestyle= 'dashed',linewidth=2)

plt.text(0.30, 0.55,f'mean:{np.mean(x):.2F}',fontsize="large",fontstyle='normal',fontweight='bold')
plt.text(0.30, 0.52,f'median:{np.median(x):.2F}',fontsize="large",fontstyle='normal',fontweight='bold')
plt.text(0.30,0.49,f'mode:{mode[0][0]:.2F}', fontsize="large", fontstyle='normal', fontweight='bold')
plt.legend(loc='upper right', bbox_to_anchor=(1.1,1))
plt.savefig("Left_Skewed_Trace_Stats.png",bbox_inches='tight')
```

```python
#create right-skewed data
x= skewnorm.rvs(a= +5, size=1000) #Skewnorm function (negative number left skew)

#create histogram of right-skewed distributed
sns.displot(x, color='darkgray',line_kws={"color":"red"})
plt.title("Right Skewed", fontsize = 15)
plt.xlabel("Range")
plt.ylabel("Frequency")
plt.savefig("Right_Skewed.png",bbox_inches='tight')
```

```python
#create trace for right-skewed distributed data
sns.displot(x, kind='kde', color='darkgray')
plt.title("Right Skewed", fontsize = 15)
plt.xlabel("Range")
plt.savefig("Right_Skewed_Trace.png",bbox_inches='tight')
```

```python
#Trace+ Histogram of right-skewed distributed data
sns.displot(x, kde=True,color='darkgray',line_kws={"color":"red"})
plt.title("Right Skewed", fontsize = 15)
plt.xlabel("Range")
plt.ylabel("Frequency")
plt.savefig("Right_Skewed_plus_Trace.png",bbox_inches='tight')
```

```python
#Trace and CENTER statistics for right-skewed distributed data
mode= stats.mode(x)
sns.displot(x, kind='kde', color='darkgray')
plt.title("Right Skewed\n (mean > median)", fontsize = 15)
plt.xlabel("Range")
plt.axvline(x.mean(), color='red', label='mean', linewidth=2)
plt.axvline(np.median(x), color='black', label='median', linestyle= 'dashed',linewidth=2)
plt.axvline(mode[0][0], color='blue', label='mode', linestyle= 'dashed',linewidth=2)
plt.text(3.1, 0.56,f'mean:{np.mean(x):.2F}',fontsize="large",fontstyle='normal',fontweight='bold')
plt.text(3.1, 0.53,f'median:{np.median(x):.2F}',fontsize="large",fontstyle='normal',fontweight='bold')
plt.text(3.1,0.50,f'mode:{mode[0][0]:.2F}', fontsize="large", fontstyle='normal', fontweight='bold')
plt.legend()
plt.savefig("Right_Skewed_Trace_Stats.png",bbox_inches='tight')
```

```python
#Generate data for bimodal distribution (two slightly different populations)
N=1000
mu, sigma = 40, 5
mu2, sigma2 = 10, 5
X1 = np.random.normal(mu, sigma, N)
X2 = np.random.normal(mu2, sigma2, N)
x= np.concatenate([X1, X2]) #merge

#create histogram for bimodal distribution
sns.displot(x, color='darkgray')
plt.title("Bimodal", fontsize = 15)
plt.xlabel("Range")
plt.ylabel("Frequency")
plt.savefig("Bimodal.png",bbox_inches='tight')
```

```python
#create trace for bimodal distributed data
sns.displot(x, kind='kde', color='darkgray')
plt.title("Bimodoal", fontsize = 15)
plt.xlabel("Range")
plt.savefig("Bimodal_Trace.png",bbox_inches='tight')
```

```python
#Trace+ Histogram of bimodal distributed data
sns.displot(x, kde=True,color='darkgray',line_kws={"color":"red"})
plt.title("Bimodal", fontsize = 15)
plt.xlabel("Range")
plt.ylabel("Frequency")
plt.savefig("Bimodal_plus_Trace.png",bbox_inches='tight')
```

```python
#Trace and CENTER statistics for bimodal distributed data
mode= stats.mode(x)
sns.displot(x, kind='kde', color='darkgray')
plt.title("Bimodal\n (mean = median)", fontsize = 15)
plt.xlabel("Range")
plt.axvline(x.mean(), color='red', label='mean', linewidth=2)
plt.axvline(np.median(x), color='black', label='median', linestyle= 'dashed',linewidth=2)
plt.axvline(mode[0][0], color='blue', label='mode', linestyle= 'dashed',linewidth=2)
plt.text(53, 0.0275,f'mean:{np.mean(x):.2F}',fontsize="large",fontstyle='normal',fontweight='bold')
plt.text(53, 0.0260,f'median:{np.median(x):.2F}',fontsize="large",fontstyle='normal',fontweight='bold')
plt.text(53, 0.0245,f'mode:{mode[0][0]:.2F}', fontsize="large", fontstyle='normal', fontweight='bold')
plt.legend(loc='upper right', bbox_to_anchor=(1.1,1))
plt.savefig("Bimodal_Trace_Stats.png",bbox_inches='tight')
```

#### Spread
- Range
- Standard Deviation
- IQR

```python
#Data 
x = np.random.normal(0,1.5, 1000)
y = np.random.normal(0,3, 1000)
z = np.random.normal(0,0.75, 1000)

#Plot
sns.distplot(z, hist=True, color="green", label="Narrow")
sns.distplot(x, hist=True, color="blue", label="Normal")
sns.distplot(y, hist=True, color="red", label= "Wide")

plt.title("Spread (Narrow, Normal, Wide)", fontsize = 15)
plt.xlabel("Range")

plt.legend()
plt.savefig("Spread_All_Hist.png",bbox_inches='tight')
```

```python
#Plot
sns.distplot(z, hist=False, color="green", label="Narrow")
sns.distplot(x, hist=False, color="blue", label="Normal")
sns.distplot(y, hist=False, color="red", label= "Wide")

plt.title("Spread (Narrow, Normal, Wide)", fontsize = 15)
plt.xlabel("Range")

plt.legend()
plt.savefig("Spread_All_Trace.png",bbox_inches='tight')
```

```python
#Plot
sns.distplot(x, hist=True, color="blue", label="Normal")
plt.title("Spread (Normal)", fontsize = 15)
plt.xlabel("Range")
plt.text(4, 0.27,f'STD:{np.std(x):.2F}',fontsize="large",fontstyle='normal',fontweight='bold')
plt.text(4, 0.25,f'IQR:{stats.iqr(x):.2F}',fontsize="large",fontstyle='normal',fontweight='bold')
plt.text(3.6, 0.23,f'Range:{np.max(x)-np.min(x):.2F}',fontsize="large",fontstyle='normal',fontweight='bold')

plt.legend()
plt.savefig("Spread_Normal.png",bbox_inches='tight')
```

```python
#Plot
sns.distplot(z, hist=True, color="green", label="Narrow")
plt.title("Spread (Narrow)", fontsize = 15)
plt.xlabel("Range")
plt.text(1.8, 0.48,f'STD:{np.std(z):.2F}',fontsize="large",fontstyle='normal',fontweight='bold')
plt.text(1.8, .44,f'IQR:{stats.iqr(z):.2F}',fontsize="large",fontstyle='normal',fontweight='bold')
plt.text(1.8, 0.40,f'Range:{np.max(z)-np.min(z):.2F}',fontsize="large",fontstyle='normal',fontweight='bold')
plt.legend()
plt.savefig("Spread_All_Narrow.png",bbox_inches='tight')
```

```python
#Plot
sns.distplot(y, hist=True, color="red", label="Wide")
plt.title("Spread (Wide)", fontsize = 15)
plt.xlabel("Range")
plt.text(7.5, 0.125,f'STD:{np.std(y):.2F}',fontsize="large",fontstyle='normal',fontweight='bold')
plt.text(7.5, .115,f'IQR:{stats.iqr(y):.2F}',fontsize="large",fontstyle='normal',fontweight='bold')
plt.text(6.2,0.105,f'Range:{np.max(y)-np.min(y):.2F}',fontsize="large",fontstyle='normal',fontweight='bold')
plt.legend()
plt.savefig("Spread_All_Wide.png",bbox_inches='tight')
```

#### Kurtosis & Skewness

```python
#Data
a=np.random.laplace(0,1,1000)
kurt_a= kurtosis (a, bias=False, fisher=True)

#Plot
sns.distplot(a, hist=True, color="blue", label="{}, {}".format("Leptokurtic", round(kurt_a, 2)))
plt.title("Kurtosis (heavy tailed)", fontsize = 15)
plt.xlabel("Range")
plt.legend()
plt.savefig("Spread_Heavy_Tailed.png",bbox_inches='tight')
```

```python
#Data
b= np.random.normal(0,1,1000)
kurt_b= kurtosis (b, bias=False, fisher=True)

#Plot
sns.distplot(b, hist=True, color="green", label="{}, {}".format("Normal", round(kurt_b, 2)))
plt.title("Kurtosis (normal)", fontsize = 15)
plt.xlabel("Range")
plt.legend()
plt.savefig("Spread_Normal_Tails.png",bbox_inches='tight')
```

```python
#Data
c= np.random.uniform(-5,5,1000)
kurt_c= kurtosis (c, bias=False, fisher=True)

#Plot
sns.distplot(c, hist=True, color="red", label="{}, {}".format("Playkurtic", round(kurt_c, 2)))
plt.title("Kurtosis (light tailed)", fontsize = 15)
plt.xlabel("Range")
plt.legend()
plt.savefig("Spread_ligh_tailed.png",bbox_inches='tight')
```

```python
#Plot
sns.distplot(a, hist=True, kde=True, color="blue", label="{}, {}".format("Leptokurtic", round(kurt_a, 2)))
sns.distplot(b, hist=True, kde=True, color="green", label="{}, {}".format("Normal", round(kurt_b, 2)))
sns.distplot(c, hist=True,  kde=True, color="red", label="{}, {}".format("Playkurtic", round(kurt_c, 2)))


plt.title("Spread (Kurtosis)", fontsize = 15)
plt.xlabel("Range")

plt.legend()
plt.savefig("Spread_All_Tails.png",bbox_inches='tight')
```

```python
#Plot
sns.distplot(a, hist=False, kde=True, color="blue", label="{}, {}".format("Leptokurtic", round(kurt_a, 2)))
sns.distplot(b, hist=False, kde=True, color="green", label="{}, {}".format("Normal", round(kurt_b, 2)))
sns.distplot(c, hist=False,  kde=True, color="red", label="{}, {}".format("Playkurtic", round(kurt_c, 2)))


plt.title("Spread (Kurtosis)", fontsize = 15)
plt.xlabel("Range")

plt.legend()
plt.savefig("Spread_All_Tails_Trace_.png",bbox_inches='tight')
```

```python
#Code source https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.kurtosis.html 
x = np.linspace(-5, 5, 100)
ax = plt.subplot()
distnames = ['laplace', 'norm', 'uniform']

for distname in distnames:
    if distname == 'uniform':
        dist = getattr(stats, distname)(loc=-2, scale=4)
    else:
        dist = getattr(stats, distname)
    data = dist.rvs(size=1000)
    kur = kurtosis(data, fisher=True)
    y = dist.pdf(x)
    if(distname=='laplace'):
        ax.plot(x, y, label="{}, {}".format('Leptokurtic', round(kur, 3)))
    elif (distname=='norm'):
        ax.plot(x, y, label="{}, {}".format('Normal', round(kur, 3)))
    else:
        ax.plot(x, y, label="{}, {}".format('Playkurtic', round(kur, 3)))
        
    ax.legend()
```
